```python
import requests
import pandas as pd

#from keys import TMDB_API_KEY

BASE_URL = "https://api.themoviedb.org/3"

```

```python
def search_movie(term):
    url = f"{BASE_URL}/search/movie"
    params = {
        "api_key": TMDB_API_KEY,
        "query": term
    }
    response = requests.get(url, params=params)

    # Check to make sure results are actually being sent
    if response.status_code != 200:
        print("Error: ", response.status_code)
        print(response.text)
        return []


    data = response.json()

    # If something is returned that isn't expected
    if "results" not in data or data["results"] is None:
        print("TMDB returned no results.")
        print(data)
        return []
    results = data.get("results", [])

    movies = []
    for m in results:
        movies.append({
            "id": m["id"],
            "title": m["title"],
            "year": (m.get("release_date") or "")[:4],
            "overview": m.get("overview", ""),
            "genre_ids": m.get("genre_ids", [])
        })



    return movies
```

```python
# Testing search_movie with hardcoded search term
movies = search_movie("Idiocracy")
for i, m in enumerate(movies, start=1):
    print(i, m["id"], m["title"], m["year"])
```

```python
# CH: make name of review file an argument
def submit_review(movie_id, rating, genre_ids=None, reviews_csv="reviews.csv"):
    # Validate User review
    if rating < 1 or rating > 5:
        print("You must pick between 1 and 5 stars.")
        return
    
    # Load existing reviews (if any)
    try:
        df = pd.read_csv(reviews_csv)
    except FileNotFoundError:
        df = pd.DataFrame(columns=["movie_id", "rating", "genre_ids"])

    # Add a new row for the new review
    new_row = {
        "movie_id": movie_id,
        "rating": rating,
        "genre_ids": str(genre_ids) if genre_ids is not None else "" # Add empty string if no genre id string to keep csv clean
    }
    # Add new review to the bottom of reviews.csv
    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

    # Save new review to reviews.csv
    df.to_csv(reviews_csv, index=False)
    print("Your review was saved as", reviews_csv)
```

```python
def get_movie_details(movie_id):

    # Get movie details from TMDB
    url = f"{BASE_URL}/movie/{movie_id}"
    params = {
        "api_key": TMDB_API_KEY
    }
    response = requests.get(url, params=params)

    # Check to make sure request is fufilled by TMDB API
    if response.status_code != 200:
        print("TMDB request error:", response.status_code)
        return {}
    
    # Return details in JSON format
    return response.json()
```

```python
def generate_movieboard(top_n=10):
    # Read reviews.csv and check to see if there is reviews
    try:
        df = pd.read_csv("reviews.csv")
    except FileNotFoundError:
        print("No reviews yet.")
        return
    
    # No reviews present?
    if df.empty:
        print("No reviews yet.")
        return
    
    # group movies by ID
    grouped_ID = df.groupby("movie_id")["rating"].agg(["mean", "count"]).reset_index()
    grouped_ID = grouped_ID.sort_values(by=["mean", "count"], ascending=[False, False])

    # Show top movies based on set value (n)
    print("Top movies:")
    for i, row in grouped_ID.head(top_n).iterrows():
        movie_id = int(row["movie_id"])
        avg = row["mean"]
        count = int(row["count"])

        # Get movie title from TMDB 
        details = get_movie_details(movie_id)
        title = details.get("title", "Unknown title")

        # Print movieboard entry
        print(f"{title} (ID: {movie_id}) -- Avg: {avg: .2f} from {count} reviews")
    
```

```python
# Test generate_movieboard
generate_movieboard()
```

```python
# Example: Saving and reading lists in CSV files
import pandas as pd
import ast

# Create a sample DataFrame with a column containing lists
sample_data = {
    "movie_id": [101, 102, 103],
    "rating": [4.5, 3.0, 5.0],
    "genre_ids": [[28, 12, 878], [35, 10749], [18, 36]]
}
df_example = pd.DataFrame(sample_data)
print("Original DataFrame:")
print(df_example)
print("\nData types:")
print(df_example.dtypes)
```

```python

# Save to CSV - lists are automatically converted to strings
df_example.to_csv("example_lists.csv", index=False)
print("Saved to example_lists.csv")
```

```python

# Read back from CSV - genre_ids will be strings, not lists
df_read = pd.read_csv("example_lists.csv")
print("\nDataFrame read from CSV:")
print(df_read)
print("\nData types after reading:")
print(df_read.dtypes)
print(f"\nType of genre_ids[0]: {type(df_read['genre_ids'][0])}")
print(f"Value: {df_read['genre_ids'][0]}")
```

```python

# Convert string back to list using ast.literal_eval
df_read['genre_ids'] = df_read['genre_ids'].apply(ast.literal_eval)
print("\nAfter converting strings back to lists:")
print(df_read)
print(f"\nType of genre_ids[0]: {type(df_read['genre_ids'][0])}")
print(f"Value: {df_read['genre_ids'][0]}")
```
